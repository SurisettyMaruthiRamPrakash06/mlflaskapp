```python
#step14
dict1={'Customer_id':[1,2,3,4,6],'Product':['Oven','Oven','Oven','Oven','Oven']}
df1=pd.DataFrame(dict1)
df1
```

```python
#step15
dict2={'Customer_id':[2,4,5,16],'Item':['Apple','Apple','Apple','Apple']}
df2=pd.DataFrame(dict2)
df2
```

```python
#step15
inner_join_df=pd.merge(df1,df2,on='Customer_id',how='inner')
inner_join_df
```

```python
#step16
outter_join_df=pd.merge(df1,df2,on='Customer_id',how='outer')
outter_join_df
```

```python
#step17
left_join_df=pd.merge(df1,df2,on='Customer_id',how='left')
left_join_df
```

```python
#step18
right_join_df=pd.merge(df1,df2,on='Customer_id',how='right')
right_join_df
```
