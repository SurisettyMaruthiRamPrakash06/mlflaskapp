```python
import pandas as pd

# Sample data for sales transactions and customer information
sales_data = {
    'TransactionID': [1, 2, 3, 4, 5],
    'CustomerID': [101, 102, 103, 104, 101],
    'Amount': [250, 300, 400, 500, 600],
    'Date': ['2025-01-01', '2025-01-02', '2025-01-03', '2025-01-04', '2025-01-05']
} # Display Amount and Date from sales data
customer_data = {
    'CustomerID': [101, 102, 103, 104],
    'CustomerName': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [30, 35, 40, 25],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}# Display Customer name and city from customer data
# Display customer ids  with amount>400
# Convert data into DataFrame
sales_df = pd.DataFrame(sales_data)
customers_df = pd.DataFrame(customer_data)
# Show basic structure of sales_df
print("Sales DataFrame:")
print(sales_df.head())

# 1. Exploring the dataset (using shape and describe)
print("\nShape of sales data:", sales_df.shape)  # Get number of rows and columns
print("\nSales data statistics:")
print(sales_df.describe())  # Summary statistics
print("Sales dataframe",sales_df)
print("customers_df",customers_df)

#   Filter a dataframe of sales dataframe where sales amount >400
#display(data.loc[(data.Brand == 'Maruti') & (data.Year == 2012)])
```

```python
# 2. Merging data (Sales with Customer info)
merged_inner = pd.merge(sales_df, customers_df, on='CustomerID', how='inner')
print("\nInner Join DataFrame:")
print(merged_inner)
```

```python
merged_outer = pd.merge(sales_df, customers_df, on='CustomerID', how='outer')
print("\nOuter Join DataFrame:")
print(merged_outer)
```

```python
merged_left = pd.merge(sales_df, customers_df, on='CustomerID', how='left')
print("\nLeft Join DataFrame:")
print(merged_left)
```

```python
merged_right = pd.merge(sales_df, customers_df, on='CustomerID', how='right')
print("\nRight Join DataFrame:")
print(merged_right)
```

```python
# Merging DataFrames to get customer name and city
merged_df = pd.merge(sales_df, customer_df, on='CustomerID')

# Displaying customer name and city from customer data
customer_info_df = merged_df[['CustomerName', 'City']]
print("Customer Name and City:")
print(customer_info_df)

# Displaying customer IDs with amount > 400
high_amount_df = sales_df[sales_df['Amount'] > 400][['CustomerID', 'Amount']]
print("\nCustomer IDs with Amount > 400:")
print(high_amount_df)
```
